```python
best_accuracy = 100
best_random_state = None

for random_state in range(1,20):
    model = RandomForestRegressor(random_state=random_state)
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    accuracy = np.sqrt(mean_squared_error(y_test, pred))
    
    if accuracy < best_accuracy:
        best_accuracy = accuracy
        best_random_state = random_state

print(f"최고 정확도: {best_accuracy:.4f}, 최적의 random state: {best_random_state}")
```

```python
train_x = train.drop("age", axis = 1)
train_y = train['age']

#Random Forest Classifier
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size=0.3,random_state=42)
model = RandomForestRegressor()
model.fit(x_train, y_train)
pred = model.predict(x_test)
accuracy_RFC = np.sqrt(mean_squared_error(y_test, pred))

#Support Vector Machines
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size=0.3,random_state=42)
model = SVR()
model.fit(x_train, y_train)
pred = model.predict(x_test)
accuracy_SVM = np.sqrt(mean_squared_error(y_test, pred))

#Gaussian Naive Bayes
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size=0.3,random_state=42)
model = GaussianNB()
model.fit(x_train, y_train)
pred = model.predict(x_test)
accuracy_NB = np.sqrt(mean_squared_error(y_test, pred))

#Decision Tree
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size=0.3,random_state=42)
model = DecisionTreeRegressor()
model.fit(x_train, y_train)
pred = model.predict(x_test)
accuracy_DT = np.sqrt(mean_squared_error(y_test, pred))

#AdaBoost
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size=0.3,random_state=42)
model = AdaBoostRegressor()
model.fit(x_train, y_train)
pred = model.predict(x_test)
accuracy_Ada = np.sqrt(mean_squared_error(y_test, pred))

#Gradient Boosting Classifier
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size=0.3,random_state=42)
model = GradientBoostingRegressor()
model.fit(x_train, y_train)
pred = model.predict(x_test)
accuracy_GBC = np.sqrt(mean_squared_error(y_test, pred))

models2 = pd.DataFrame({
    'Model': ['Random Forest', 'Support Vector Machines', 'Naive Bayes',
			  'Decision Tree', 'AdaBoostClassifier', 'GradientBoostingClassifier'],
    'RMSE': [accuracy_RFC, accuracy_SVM, accuracy_NB, 
              accuracy_DT, accuracy_Ada, accuracy_GBC]})
print(models2.sort_values(by='RMSE', ascending=True))
```
