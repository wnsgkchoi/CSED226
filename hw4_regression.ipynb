```python
# 파일 읽기
train = pd.read_csv('age_train.csv')
test = pd.read_csv('age_test.csv')

# 제출 형식에 따라 ID 추가
train['ID'] = range(0, len(train))
test['ID'] = range(0, len(test))

# 결과값 넣을 set 준비
result = test['ID']

train = train.drop(['ID'], axis = 1)
test = test.drop(['ID'], axis = 1)

#복사본 생성
original_train = train.copy()
full_data = [train, test]
```

```python

#성을 0,1,2로 변환
train['gender'] = train['gender'].map({'M' : 0, 'F' : 1, 'I' : 2})
test['gender'] = test['gender'].map({'M' : 0, 'F' : 1, 'I' : 2})

# gender와 age의 관계를 살펴보면 8세 이하일수록 I가 많고, 9~11세 사이는 M이 많고, 그보다 나이가 많은 경우 M이나 F일 확률이 높다.
# sns.countplot(data=train, x='age', hue='gender')
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_train_s, y_train, random_state=30)
K = 50
acc = np.zeros([K])

for n in range(1, K+1):
    # Create a KNeighborsRegressor with n_neighbors set to 'n'
    regressor = KNeighborsRegressor(n_neighbors=n)
    regressor.fit(x_train, y_train.ravel())
    y_pred = regressor.predict(x_test)
    # Calculate the mean squared error as the accuracy metric
    acc[n-1] = metrics.mean_squared_error(y_test, y_pred)

    print(acc)
print(f'Minimum Mean Squared Error: {np.min(acc)}')
print(f'Maximum Mean Squared Error: {np.max(acc)}')
min_mse_index = np.argmin(acc) + 1
print(f'N value for Minimum Mean Squared Error: {min_mse_index}')
```
