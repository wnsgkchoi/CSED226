```python


# 데이터가 이미 x_train, y_train, x_test, y_test로 나누어져 있다고 가정

best_accuracy = 0
best_random_state = None

for max_depth in range(2,20):
    model = RandomForestClassifier(random_state=2, min_samples_split=max_depth)
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    accuracy = accuracy_score(pred, y_test)
    
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_random_state = max_depth

print(f"최고 정확도: {best_accuracy:.4f}, 최적의 random state: {best_random_state}")
```

```python

params = {
           'max_depth' : list(range(1,20)),
           'min_samples_leaf' : list(range(1,20)),
           'min_samples_split' : list(range(1,20))
            }

rf_clf = RandomForestClassifier(random_state = 2, n_jobs = -1)
grid_cv = GridSearchCV(rf_clf, param_grid = params, cv = 3, n_jobs = -1)
grid_cv.fit(x_train, y_train)

print('최적 하이퍼 파라미터: ', grid_cv.best_params_)
print('최고 예측 정확도: {:.4f}'.format(grid_cv.best_score_))
```

```python
train_x = train.drop("Class", axis = 1)
train_y = train['Class']

#Random Forest Classifier
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size=0.3,random_state=42)
model = RandomForestClassifier()
model.fit(x_train, y_train)
pred = model.predict(x_test)
accuracy_RFC = accuracy_score(pred,y_test)
result_RFC=cross_val_score(model,train_x,train_y,cv=10,scoring='accuracy')


#Support Vector Machines
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size=0.3,random_state=42)
model = SVC()
model.fit(x_train, y_train)
pred = model.predict(x_test)
accuracy_SVM = accuracy_score(pred,y_test)
result_SVM=cross_val_score(model,train_x,train_y,cv=10,scoring='accuracy')

#Gaussian Naive Bayes
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size=0.3,random_state=42)
model = GaussianNB()
model.fit(x_train, y_train)
pred = model.predict(x_test)
accuracy_NB = accuracy_score(pred,y_test)
result_NB =cross_val_score(model,train_x,train_y,cv=10,scoring='accuracy')

#Decision Tree
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size=0.3,random_state=42)
model = DecisionTreeClassifier()
model.fit(x_train, y_train)
pred = model.predict(x_test)
accuracy_DT = accuracy_score(pred,y_test)
result_DT =cross_val_score(model,train_x,train_y,cv=10,scoring='accuracy')

#AdaBoost
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size=0.3,random_state=42)
model = AdaBoostClassifier()
model.fit(x_train, y_train)
pred = model.predict(x_test)
accuracy_Ada = accuracy_score(pred,y_test)
result_Ada =cross_val_score(model,train_x,train_y,cv=10,scoring='accuracy')

#Gradient Boosting Classifier
x_train,x_test,y_train,y_test = train_test_split(train_x,train_y,test_size=0.3,random_state=42)
model = GradientBoostingClassifier()
model.fit(x_train, y_train)
pred = model.predict(x_test)
accuracy_GBC = accuracy_score(pred,y_test)
result_GBC =cross_val_score(model,train_x,train_y,cv=10,scoring='accuracy')

models = pd.DataFrame({
    'Model': ['Random Forest', 'Support Vector Machines', 'Naive Bayes',
			  'Decision Tree', 'AdaBoostClassifier', 'GradientBoostingClassifier'],
    'Score': [result_RFC.mean(), result_SVM.mean(), result_NB.mean(), 
              result_DT.mean(), result_Ada.mean(), result_GBC.mean()]})
print(models.sort_values(by='Score',ascending=False))

models2 = pd.DataFrame({
    'Model': ['Random Forest', 'Support Vector Machines', 'Naive Bayes',
			  'Decision Tree', 'AdaBoostClassifier', 'GradientBoostingClassifier'],
    'Score': [accuracy_RFC, accuracy_SVM, accuracy_NB, 
              accuracy_DT, accuracy_Ada, accuracy_GBC]})
print(models2.sort_values(by='Score',ascending=False))
```
