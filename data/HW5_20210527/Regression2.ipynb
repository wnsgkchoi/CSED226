```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("bmh")
from sklearn import preprocessing
from sklearn.svm import SVR
from sklearn import utils
from sklearn.tree import DecisionTreeRegressor


# model
from sklearn.ensemble import RandomForestRegressor
import optuna
from sklearn.ensemble import BaggingRegressor

from sklearn.linear_model import Lasso
# misc
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import metrics


data=pd.read_csv("train_regression.csv")
data.head()



col=['acid0','acid1','acid2','chlorides','sulphates0','sulphates1','sulphates2','density']
X=data.iloc[:, :-1]
y=data['score']
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=27)



model = BaggingRegressor(
                                base_estimator= DecisionTreeRegressor(max_depth=140),
                                 n_estimators = 200
                                 ,
                                 max_samples = 1.0,
                                 max_features = 1.0
                        ) 

model.fit(X_train, y_train) 
accuracy = sqrt(mean_squared_error(model.predict(X_test), y_test))


accuracy
```
