```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
rating = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
```

```python
from surprise import Dataset
from surprise import Reader
 
min_rating = rating.rating.min()
max_rating = rating.rating.max()
 
reader = Reader(rating_scale=(min_rating, max_rating))
data = Dataset.load_from_df(rating[['user_id', 'item_id', 'rating']], reader)
```

```python
from surprise import SVD
from surprise.model_selection import cross_validate
 
svd = SVD(n_epochs=10)
results = cross_validate(svd, data, measures=['RMSE', 'MAE'], cv=10, verbose=True)
```

```python
from surprise.prediction_algorithms import KNNBasic
knn = KNNBasic()
```

```python
print("Average MAE: ", np.average(results["test_mae"]))
print("Average RMSE: ", np.average(results["test_rmse"]))
```

```python
from surprise import SVD
from surprise.model_selection import GridSearchCV
 
param_grid = {
  'n_factors': [18,19,20,21,22,23,24],
  'n_epochs': [24,25,26,27,28,29,30]
}
 
gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=10)
gs.fit(data)
 
print(gs.best_score['rmse'])
print(gs.best_params['rmse'])
```

```python
from surprise.model_selection import train_test_split
from surprise import accuracy, Dataset, SVD
best_factor = gs.best_params['rmse']['n_factors']
best_epoch = gs.best_params['rmse']['n_epochs']
trainset, testset = train_test_split(data, test_size=.20)
finalmodel = SVD(n_factors=best_factor, n_epochs=best_epoch)



```
